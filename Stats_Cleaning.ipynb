### Dealing with null Values

```python
# checking the mean of the age column
df["age"].mean()
```

```python
df.isnull().sum()
```

```python
# filling with the below values of null values in age column
df.groupby('marital')['age'].median()
```

```python
new_age = []

for i in df.values:
    if type(i[1]) == float:
        if i[4] == "single":
            new_age.append(32)
        elif i[4] == "married":
            new_age.append(42)
        else:
            new_age.append(45)
    else:
        new_age.append(i[1])
```

```python
len(new_age)
```

```python
df["age"] = new_age
```

```python

# after filling checking the age column mean 
df["age"].mean()
```

```python
# replacing question mark with np.nan
new_salary = []

for i in df.salary:
    if type(i) == int:
        new_salary.append(i)
    else:
        new_salary.append(np.nan)
len(new_salary)   
df["salary"] = new_salary
```

```python
df.isnull().sum()
```

```python
df["salary"].fillna(0, inplace = True)
```

```python
df.isnull().sum()
```

```python
# filling null values with bfill
df.month.bfill(inplace = True)
```

```python

df.isnull().sum()
```

```python
# whole dataframe is free from null values
df.dropna(axis = 0, how ='any', inplace = True) 
```

```python
df.isnull().sum()
```

### Seperating columns from one column

```python
edu = []
job = []
for i in df.jobedu:
    job.append(i.split(",")[0])
    edu.append(i.split(",")[1])
    
df["Education"] = edu
df["Job"] = job
```

```python
new = []
for i in df.month:
    new.append(i.split(",")[0])
print(len(new))
df["new_month"] = new
```

```python
new_duration = []
for i in df.duration:
    x = (i.split())
    #print(x)
    if x[0].isalnum() and x[1] == "sec":
        #print("seconds")
        sec = i.split(" ")[0]
        #print(sec)
        h1 = math.ceil((float(sec)/60))
        new_duration.append(h1)
        #pass
    else:
        num2 = float(i.split(" ")[0])
        h2 = math.ceil(num2)
        new_duration.append(h2)
        
print(len(new_duration))
df["duration"] = new_duration
```

```python
# droping the jobedu columns after extraing data from it 
df.drop(["jobedu"], axis = 1, inplace = True)
```

```python
df["full_date"] = df["day"].astype(str)+ "," + df["month"]
```
